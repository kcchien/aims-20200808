````python
%matplotlib inline
import matplotlib.pyplot as plt # 用matplotlib將圖檔顯示在Jupyter notebook裡
from PIL import Image
import pytesseract
````

## 中文圖檔範例

````python
img = Image.open("tra_text.jpg")
#img.show()
plt.imshow(img)
plt.show()
````

## 將含有文字的圖檔，轉換成黑白影像

為了提升OCR的效率，一律將圖檔轉成黑白圖檔，我們可以使用 PIL套件的 convert方法來完成

PIL.image.convert 的詳細使用可以參照官方文件

https://pillow.readthedocs.io/en/latest/reference/Image.html#PIL.Image.Image.convert

https://pillow.readthedocs.io/en/latest/handbook/concepts.html#concept-modes

````python

img = img.convert("L") # (8-bit pixels, black and white)
plt.imshow(img)
plt.show()
````

## 進行中文 OCR 辨識

````python
ans = pytesseract.image_to_string(img, lang="chi_tra")
print(ans)
````

## 英文圖檔範例

````python
img = Image.open("en_text.jpg")
#img.show()
plt.imshow(img)
plt.show()
````

````python
img = img.convert("L") # (8-bit pixels, black and white)
plt.imshow(img)
plt.show()
````
